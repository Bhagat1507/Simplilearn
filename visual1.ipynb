```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("auto_data.csv")
```

```python
data.head()
```

```python
def replace(num):
    if num==1:
        return "USA"
    elif num==2:
        return "France"
    elif num==3:
        return "Asia"
```

```python
data["origin"]=data["origin"].apply(replace)
```

```python
data.tail()
```

```python
sns.pairplot(data[['mpg','weight','origin']],hue='origin',size=4)
```
