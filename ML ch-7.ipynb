```python
from PIL import Image
from sklearn.cluster import KMeans
```

```python
img=Image.open('dogs.jpeg')
```

```python
from matplotlib import pyplot as plot
plot.imshow(img)
```

```python
import numpy as np
```

```python
img_np=np.asarray(img)
```

```python
img_np.shape
```

```python
img_np=img_np.reshape(185*272,3)
```

```python
img_np.shape
```

```python
km=KMeans(n_clusters=3,random_state=0)
km.fit(img_np)
```

```python
labels=km.predict(img_np)
labels=labels.reshape((185,272))
```

```python
plot.imshow(labels)
```

```python
centers=[centers.astype(int) for centers in km.cluster_centers_]
```

```python
blob=np.ones((100,100,3))
```

```python
blob1 = blob * centers[0]
plot.imshow((blob1*255).astype(np.uint8))
```

```python
blob2 = blob * centers[1]
plot.imshow((blob2*255).astype(np.uint8))
```

```python
blob3 = blob * centers[2]
plot.imshow((blob3*255).astype(np.uint8))
```
